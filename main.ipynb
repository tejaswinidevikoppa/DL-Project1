```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_march = pd.read_parquet('yellow_tripdata_2020-03.parquet')
df_may = pd.read_parquet('yellow_tripdata_2020-05.parquet')
```

```python
df_march.isnull().sum()
```

```python
df_march.dtypes
```

```python
df_march['passenger_count'].count()
```

```python
2969879/37808
```

```python
df_march['passenger_count'].fillna(df_march['passenger_count'].median(), inplace=True)
df_march['RatecodeID'].fillna(df_march['RatecodeID'].mode()[0], inplace=True)
df_march['store_and_fwd_flag'].fillna(df_march['store_and_fwd_flag'].mode()[0], inplace=True)
df_march['congestion_surcharge'].fillna(df_march['congestion_surcharge'].median(), inplace=True)
```

```python
df_march.isnull().sum()
```

```python
df_march
```

```python
df_may.isnull().sum()
```

```python
df_may['passenger_count'].fillna(df_may['passenger_count'].median(), inplace=True)
df_may['RatecodeID'].fillna(df_may['RatecodeID'].mode()[0], inplace=True)
df_may['store_and_fwd_flag'].fillna(df_may['store_and_fwd_flag'].mode()[0], inplace=True)
df_may['congestion_surcharge'].fillna(df_may['congestion_surcharge'].median(), inplace=True)

df_may.drop('airport_fee', axis=1, inplace=True)
```

```python
df_may.isnull().sum()
```

```python
#Performing EDA on data with Pandas, Numpy libraries
print("Summary Statistics for March Data:\n", df_march.describe())
print("Summary Statistics for May Data:\n", df_may.describe())
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_march['passenger_count'], bins=range(1, 8), alpha=0.5, label='March', edgecolor='black')
plt.hist(df_may['passenger_count'], bins=range(1, 8), alpha=0.5, label='May', edgecolor='black')
plt.title('Histogram of Passenger Count')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.xticks(range(1, 8))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_march['trip_distance'], bins=30, alpha=0.5, label='March')
plt.hist(df_may['trip_distance'], bins=30, alpha=0.5, label='May')
plt.title('Histogram of Trip Distance')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_march['fare_amount'], bins=30, alpha=0.5, color='orange', label='March')
plt.hist(df_may['fare_amount'], bins=30, alpha=0.5, color='red', label='May')
plt.title('Histogram of Fare Amount')
plt.xlabel('Fare Amount ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_cols_march = df_march.select_dtypes(include=['float64', 'int64'])
numeric_cols_may = df_may.select_dtypes(include=['float64', 'int64'])

correlation_march = numeric_cols_march.corr()
correlation_may = numeric_cols_may.corr()

plt.figure(figsize=(12, 6))

# March correlation heatmap
sns.heatmap(correlation_march, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap - March Data')


```

```python
# May correlation heatmap
plt.figure(figsize=(12, 6))

sns.heatmap(correlation_may, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap - May Data')

plt.tight_layout()
plt.show()
```
